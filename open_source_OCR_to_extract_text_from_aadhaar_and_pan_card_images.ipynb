```python
import cv2
import pytesseract
import numpy as np
import re
from PIL import Image

# Set Tesseract path (NOT needed in Colab, and Windows-specific path removed)
# pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

def preprocess_image(image_path):
    # Read image
    img = cv2.imread(image_path)

    # Check if image was loaded successfully
    if img is None:
        print(f"Error: Could not read image from {image_path}")
        return None

    # Resize for better OCR accuracy
    img = cv2.resize(img, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Remove noise
    gray = cv2.medianBlur(gray, 3)

    # Apply thresholding
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

    return thresh


def extract_text(image_path):
    processed_img = preprocess_image(image_path)

    if processed_img is None:
        return ""

    # OCR extraction
    text = pytesseract.image_to_string(processed_img, lang='eng')

    return text


def extract_aadhaar_details(text):
    aadhaar_number = re.findall(r"\d{4}\s\d{4}\s\d{4}", text)
    return aadhaar_number


def extract_pan_details(text):
    pan_number = re.findall(r"[A-Z]{5}[0-9]{4}[A-Z]{1}", text)
    return pan_number


if __name__ == "__main__":
    # Use the globally defined image_path
    # image_path = "card_image.jpg" # This line is no longer needed

    text = extract_text(image_path)

    print("---- Extracted Raw Text ----")
    print(text)

    # Aadhaar extraction
    aadhaar = extract_aadhaar_details(text)
    if aadhaar:
        print("Aadhaar Number:", aadhaar[0])

    # PAN extraction
    pan = extract_pan_details(text)
    if pan:
        print("PAN Number:", pan[0])
```
